```python
def cleaned_mean(arr):
    if arr==None:
        m=0
    else:
        arr = list(filter(lambda x: x!= ['ooa'], arr))
        if [] in arr:
            arr.remove([])
        df = pd.DataFrame(arr,columns=['value'])
        max_treshold=df['value'].quantile(0.80)
        df=df[df['value']<max_treshold]
        df=df[df['value']>0]
        m=np.mean(df.value)
    return m
```

```python
mn=cleaned_mean(p)
print(mn)
```

```python
def remove_ooa(arr,m):
    count=0
    array=[]
    if arr==None:
        array=None
    else:
        for i in range(len(arr)):
            if type(arr[i]) == type([]):
                if arr[i]==[]:
                    array.append(m)
                elif arr[i]==['ooa']:
                    count+=1
                    array.append(m)
            else:
                array.append(arr[i])
    return array,count
            
```

```python
def clean(x):
    mod_x=[]
    for i in x:
        array=[]
        for j in i:
            mn=cleaned_mean(j)
            s,_=remove_ooa(j,mn)
            array.append(s)
        mod_x.append(array)
    return mod_x
```

```python
new_sT=clean(sT)
new_sT
```

```python
def reliability_matrix(arr):
    rel_matrix=[]
    for i in range(len(arr)):
        array=[]
        for j in range(len(arr[i])):
            if arr[i][j]==None:
                s=None
            else:
                df = pd.DataFrame(arr[i][j],columns=['value'])
                print(i,j)
                outliers = setup(df ,silent=True,numeric_features = ["value"])
                knn = create_model('knn')
                knn_results = assign_model(knn)
                s=np.array(knn_results.Anomaly_Score)
                s=1-s
                s=np.round(100*s,2)
                s=[0 if x==100 else x for x in s]
            array.append(s)
        rel_matrix.append(array)
    return rel_matrix
```

```python
rel_matrix=reliability_matrix(new_sT)
```
